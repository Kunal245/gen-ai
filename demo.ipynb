```python
from langchain_community.document_loaders import TextLoader
```

```python
loader = TextLoader("dataset.txt", encoding="utf-8")
```

```python
text_data = loader.load()
```

```python
print(text_data[0])
```

```python
#Embedding using Hugging Face Model
```

```python
pip install sentence-transformers
```

```python
pip install langchain-huggingface
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model="all-MiniLM-L6-v2")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
textSplitter = RecursiveCharacterTextSplitter(chunk_size = 200, chunk_overlap = 20)
finalDocument = textSplitter.split_documents(text_data)
print(finalDocument)
```

```python
!pip install faiss-cpu
# pip install faiss-cpu
# !pip install faiss-gpu
```

```python
from langchain_community.vectorstores import FAISS
db = FAISS.from_documents(finalDocument, embeddings)
db
```

```python
query = input("Enter your questions for C++: ")
docs = db.similarity_search(query)
for i in range(len(docs)) :
    print(docs[i].page_content)
```
